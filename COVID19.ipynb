```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("covid.csv")

# Filter by country
country = 'USA'
country_df = df[df['Country'] == country]

# Convert Date
country_df['Date'] = pd.to_datetime(country_df['Date'])

# Plot Cases over Time
plt.figure(figsize=(10, 5))
plt.plot(country_df['Date'], country_df['Cases'], marker='o', label='Cases', color='orange')
plt.plot(country_df['Date'], country_df['Deaths'], marker='x', label='Deaths', color='red')
plt.title(f'COVID-19 Trend in {country}')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("covid_trend.png")
plt.show()
```
